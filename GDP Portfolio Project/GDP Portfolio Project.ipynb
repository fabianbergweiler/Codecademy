__Looking closer at GDP__

```python
print(df.describe())
```

```python
print(df.GDP.mean())
```

```python
plt.hist(df.GDP)
plt.ylabel('Frequency Count')
plt.xlabel('GDP in trillion US dollars')
plt.title('Distribution of GDP')
plt.axvline(df.GDP.mean(), color='red', linestyle='-')
plt.legend(['GDP', 'Mean'])
plt.show()
plt.clf()
```

GDP per country

```python
chile = df.loc['Chile']
china = df.loc['China']
germany = df.loc['Germany']
mexico = df.loc['Mexico']
usa = df.loc['United States of America']
zimbabwe = df.loc['Zimbabwe']
year = df.Year[0:16]

plt.plot(year, chile.GDP)
plt.plot(year, china.GDP)
plt.plot(year, germany.GDP)
plt.plot(year, mexico.GDP)
plt.plot(year, usa.GDP)
plt.plot(year, zimbabwe.GDP)
plt.xlabel('Year')
plt.ylabel('GDP (in trillion USD)')
plt.title('GDP Development')
plt.axhline(df.GDP.mean(), linestyle='--')
plt.legend(['Chile', 'China', 'Germany', 'Mexico', 'United States of America', 'Zimbabwe', 'Mean'])
plt.show()
```
